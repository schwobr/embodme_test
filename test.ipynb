```python
import threading
import time
import struct
import os
import numpy as np
import cv2
import argparse
import matplotlib.pyplot as plt
```

```python
def show_images(left_images, right_images):
    for k in range(left_images.shape[0]):
        cv2.imshow('Frame L', left_images[k])
        cv2.imshow('Frame R', right_images[k])

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break;

    cv2.destroyAllWindows()
```

On charge les images comme dans show_images.py, j'ai supprim√© la variable args car elle me provoquait une erreur et me semblait inutile ici.

```python
path = "raw_np"

list_img_file = os.listdir(path)
list_img_file.sort(key=lambda f: int(''.join(filter(str.isdigit, f))))
left_images = []
right_images = []

for index, img_file in enumerate(list_img_file):

    print(img_file)

    leftRightImage = np.load(os.path.join(path, img_file))
    left_images.append(leftRightImage[0])
    right_images.append(leftRightImage[1])

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break;

cv2.destroyAllWindows()
```

On met tout √ßa dans des tableaux numpy

```python
left_images = np.array(left_images)
right_images = np.array(right_images)
```

Premier essai avec un thresholding basique, sachant que la main est relativement lumineuse

```python
left_binary = np.zeros(left_images.shape)
right_binary = np.zeros(right_images.shape)
left_binary[left_images>50] = 255
right_binary[right_images>50] = 255
```

On constate 2 d√©faut principaux :
* La main disparait par endroits, notamment lorsqu'elle sort du centre de l'image (zone la plus √©clair√©e), ou lorsqu'elle s'√©loigne (o√π elleest naturellement moins √©clair√©e)
* Des zones constantes tr√®s lumineuses sont visibles sur les bords de l'image. On devrait pouvoir passer outre tant que la main est bien d√©limit√©e, on supprimera tous les √©l√©ments non centraux

```python
show_images(left_binary, right_binary)
```

Ici on va essayer une 2eme m√©thode avec la fonction threshold d'opencv. On commence par appliquer un flou gaussien √† l'image pour supprimer les imperfections aux contours, puis on applique un thresholding.

```python
left_blur = np.array([cv2.GaussianBlur(img, (5, 5), 0) for img in left_images])
right_blur = np.array([cv2.GaussianBlur(img, (5, 5), 0) for img in right_images])
print(left_blur.shape)
```

```python
left_binary = np.array([cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1] for img in left_blur])
right_binary = np.array([cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1] for img in right_blur])
```

Ici le r√©sultat est bien meilleur, la main a d√©j√† moins tendance √† dispara√Ætre et semble plus nette

```python
show_images(left_binary, right_binary)
```
